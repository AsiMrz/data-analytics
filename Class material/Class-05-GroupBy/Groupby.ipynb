--- 

**Groupby**

Here you have an easy way to get your answer because their is only 2 Sex in your dataframe. But imagine you would have 15 different types in your dataframe that you would want the `mean()` from? It would be a bit hectic to go through each on of them individually. That's were the `groupby()` method becomes interesting.




**`groupby()`** enables you to specify a column that you would want to split into categories and apply to each category an aggregation method of your choice.

The function will:

* **Split the data** into groups based on some criteria.
* **Apply a function** to each group independently.
* **Combine the results** into a data structure.


Let's solve the same exercise as before using the `groupby()` method

````
df.groupby("Embarked")[["Age"]].mean()
# df.groupby('Sex')[['Age']].mean()
````

We can specify which column we want to analyze after the grouping. In this case we only chose "Age". In the these square brackets we can add more columns as well.

````
df.groupby('Sex')[['Age']].mean()
````

### Extracting a group

Groupby gives us an awesome feature to extract a group. With ```.get_group()``` we can extract a group from the dataset and analyze it on it's own. 

````
df.groupby("Sex").get_group("female").head()
#df[df.Embarked == "C"]
````

````
df.groupby('Sex').get_group('female').head()
````

````
df.groupby("Embarked").get_group("C").describe()[["Age"]]
````

### Agg() method

The `agg()` method allows you to perform specific aggregation(s) on specific columns.



You can pass to the method 
*   **A list of aggregation methods** that you would want to perform
*   **A dictionnary** with name of the column as keys and list of aggregation methods as values



````
df.groupby('Sex').agg(["count","mean","max","min"])["Age"]
````

````
df.groupby('Sex').max()['Age']
````

````
df.groupby('Sex').median()['Age']
````

````
df.groupby('Sex').agg(['count','max','median'])[['Age','SibSp']]
````
